```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.layouts import row
import numpy as np
import rich.traceback

output_notebook()
```

```python
rich.traceback.install()
x = []
y = []
Py = []
Px = []
with open("PLAs.txt") as f:
    list = f.read().strip().split()
    for recs in list:
        y.append(recs)
    y.reverse()

with open("POL.txt") as f:
    list = f.read().strip().split()
    for recs in list:
        Py.append(recs)
    Py.reverse()
    
with open("AfgDates.txt") as f:
    list = f.read().strip().split()
    for recs in list:
        x.append(recs)
    x.reverse()


with open("PolDates.txt") as f:
    list = f.read().strip().split()
    for recs in list:
        Px.append(recs)
    Px.reverse()

dic = {'x' : x, 'y' : y}
dicp = {'Px' : Px, 'Py' : Py}
print(x)
print(Px)
```

```python
fig1 = figure(x_axis_type = 'datetime', x_range = (50, 340), y_range = (0, 30), x_axis_label = 'Dni od 31.12.2019', y_axis_label = 'Zachorowa na 100tys mieszkacw', title = 'Wykres zachorowa na COVID-19 w Afganistanie midzy 31.12.19-14.12.20')
fig1.toolbar.logo = None
fig1.toolbar.autohide = True
fig1.line(x, y, line_color = 'red')

fig2 = figure(x_range = (0, 284), y_range = (0, 1000), x_axis_label = 'Dni od 04.03.2020', y_axis_label = 'Zachorowa na 100tys mieszkacw', title = 'Wykres zachorowa na COVID-19 w Polsce midzy 31.12.19-14.12.20')
fig2.toolbar.logo = None
fig2.toolbar.autohide = True
fig2.line(Px, Py, line_color = 'red')

show(row(fig1, fig2))
```
